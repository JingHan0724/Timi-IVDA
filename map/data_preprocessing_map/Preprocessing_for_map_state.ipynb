```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("new_data.csv")
```

```python
df
```

```python
#calculate total number of people
for i in list(df.columns.values)[3:-1]:
    df["total_{}".format(i)] = df["Population2010"] * df[i]
```

```python
#drop old percentage and useless columns
drop_list = list(df.columns.values)[4:33]
```

```python
df.drop(drop_list,inplace = True, axis = 1)
```

```python
df.drop(["PlaceName","PlaceFIPS","total_Population2010"],inplace = True, axis = 1)
```

```python
df
```

```python
df2 = df.groupby("StateAbbr").sum()
```

```python
df2
```

```python
#calculate percentage by state
for i in list(df2.columns.values)[1:]:
    df2["final_{}".format(i)] = df2[i]/df2["Population2010"]
```

```python
drop_list_df2 = df2.columns.values[1:29]
```

```python
#drop number of people columns
df2.drop(drop_list_df2,inplace = True, axis = 1)
```
